```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import yankee data
yankeesbatting2024 = pd.read_csv("Yankee Batting 2024 .csv")
yankeesbatting2024['Season'] = 2024
yankeesbatting2024['Team'] = "NYY" 
yankeespitching2024 = pd.read_csv("Yankee Pitching 2024.csv")
yankeespitching2024['Season'] = 2024
yankeespitching2024['Team'] = "NYY" 

yankees2024season = pd.concat([yankeespitching2024,yankeesbatting2024])
yankees2024seasontotal = yankees2024season[['Pos','WAR',]].groupby('Pos').sum()
ax=yankees2024seasontotal.plot.bar()
ax.set_ylabel("Count")
ax.set_xlabel("2024 Position WAR")

yankeesbatting2023 = pd.read_csv("Yankee Batting 2023.csv")
yankeesbatting2023['Season'] = 2023
yankeesbatting2023['Team'] = "NYY" 
yankeespitching2023 = pd.read_csv("Yankees Pitching 2023.csv")
yankeespitching2023['Season'] = 2023
yankeespitching2023['Team'] = "NYY" 

yankees2023season = pd.concat([yankeespitching2023,yankeesbatting2023])
yankees2023seasontotal = yankees2023season[['Pos','WAR']].groupby('Pos').sum()
ax=yankees2023seasontotal.plot.bar()
ax.set_ylabel("Count")
ax.set_xlabel("2023 Position WAR")

yankeesbatting2022 = pd.read_csv("Yankee Batting 2022.csv")
yankeesbatting2022['Season'] = 2022
yankeesbatting2022['Team'] = "NYY" 
yankeespitching2022 = pd.read_csv("Yankee Pitching 2022.csv")
yankeespitching2022['Season'] = 2022
yankeespitching2022['Team'] = "NYY" 

yankees2022season = pd.concat([yankeespitching2022,yankeesbatting2022])
yankees2022seasontotal = yankees2022season[['Pos','WAR']].groupby('Pos').sum()
ax=yankees2022seasontotal.plot.bar()
ax.set_ylabel("Count")
ax.set_xlabel("2022 Position WAR")
```
